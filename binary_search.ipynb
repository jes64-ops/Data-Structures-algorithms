```python
A = [-4 , -2, 0 ,1,5,7]

#naive O(n )searching

if 7 in A :
  print(True)
```

```python
def binary_search(arr, target):
  N = len(arr)
  L = 0 ## firt item in array
  R = N-1 ##last item

  while L <= R:
    M = L + ((R-L)//2)
    if arr[M] == target:
      return True
    elif target < arr[M]:
      R = M-1
    else:
      L = M+1
  return False

binary_search(A , 1)## check if 1 is in the array
```

```python
#condition based
B =[ False,  False, False, True, True, True]

def binary_search_condition(arr):
  N = len(arr)
  L=0
  R= N-1

  while L < R:
    M = (L+ R)//2
    if B[M ]:
      R = M
    else:
      L = M+ 1

  return L

binary_search_condition(B)
```
