```python
A = [ 1, 2, 3 ]
print(A)
```

```python
# Append - Insert element at end of array - On average: O(1)

A. append(6)

print(A)
```

```python
# Pop - Deleting element at end of array - O(1)

A.pop()
print(A)
```

```python
# Insert (not at end of array) - O(n)

A. insert( 2, 5)
print(A)
```

```python
# Modify an element 0(1)
A[0] = 9
print(A)
```

```python
#access to element given index i O(1)
print(A[2])
```
