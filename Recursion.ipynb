```python
#Fibonacci
#F(0)=0
#F(1)=1
#n>1 : F(n) = F(n-1)+ F(n-2)

#time O(2^N) space : O(n)
def F(n):
  if n == 0:
    return 0
  elif n== 1:
    return 1
  else:
    return F(n- 1)+ F(n-2)

F(10)
```

```python

# singly linked list

class SinglyNode:

  def __init__(self,val, next=None):
    self.val = val
    self.next = next

  def __str__(self):
    return str(self.val)


Head = SinglyNode(5)#

A = SinglyNode(3)
B = SinglyNode(4)
C = SinglyNode(9)

Head.next = A
A.next = B
B.next = C
print(Head)# shows value of first node
```

```python
# linked lists
def reverse(node):
  if not node:
    #end of list
    return

  reverse(node.next)
  print(node)

  reverse(Head)
```
