```python
#Fibonacci
#F(0)=0
#F(1)=1
#n>1 : F(n) = F(n-1)+ F(n-2)

#time O(2^N) space : O(n)
def F(n):
  if n == 0:
    return 0
  elif n== 1:
    return 1
  else:
    return F(n- 1)+ F(n-2)

F(10)
```
