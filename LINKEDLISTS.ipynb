```python
# singly linked list

class SinglyNode:

  def __init__(self,val, next=None):
    self.val = val
    self.next = next

  def __str__(self):
    return str(self.val)
```

```python
Head = SinglyNode(5)#

A = SinglyNode(3)
B = SinglyNode(4)
C = SinglyNode(9)

Head.next = A
A.next = B
B.next = C
print(Head)# shows value of first node
```

```python
# Traverse the list - O(n)

curr = Head

while curr:
  print(curr)
  curr = curr.next
```

```python
#display the linked list  -O(n)
def display(head):
  curr = head
  elements = []
  while curr:
    elements.append(str(curr.val))
    curr = curr.next
  print('->'.join(elements))
```

```python
#search for node value = o(n)

def search(head, val):
  curr = head
  while curr:
    if curr.val == val:
      return True
    curr = curr.next
  return False

search(Head, 9)
```

```python
## Doubly linked list

class DoublyNode:
  def __init__(self, val , next=None, prev=None):
    self.val = val
    self.next = next
    self.prev = prev

  def __str__(self):
    return str(self.val)
```

```python
head = tail = DoublyNode(5)
print(tail)
```

```python
def display(head):
  curr = head
  elements =[]
  while curr:
    elements.append(str(curr.val))
    curr = curr.next
  print('<->'.join(elements))
display(head)
```

```python
## Insert at beginning _ O(1)
def insert_at_beginning(head, tail ,val):
  new_node = DoublyNode(val, next=head)
  head.prev = new_node
  return new_node , tail

head, tail = insert_at_beginning(head, tail, 4)
display(head)
```

```python
## Insert at beginning _ O(1)
def insert_at_end(head, tail ,val):
  new_node = DoublyNode(val, prev=tail)
  tail.next = new_node
  return head, new_node

head, tail = insert_at_end(head, tail, 5)
display(head)
```
