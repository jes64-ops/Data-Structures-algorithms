```python
s = set()
print(s)
```

```python
#add items to set = O(1)
s.add(2)
s.add(5)
s.add(8)
s.add(10)

print(s)
```

```python
# look up item in set

if 2 in s:
  print(True)
else:
  print(False)
```

```python
#REMOVE ITEM FROM SET - 0(1)
s.remove(5)

print(s)
```

```python
# HAshmaps = dictionaries

d = {'greg':1 , 'jesse': 2, 'eli': 3}

print(d)
```

```python
#Add key;value in dictionary : O(1)
d['daniel'] = 4
print(d)
```

```python
#check for prence of key in dictionary :
if 'jesse ' not in d:
  print(True)
else:
  print(False)
```

```python
#check the value corresponding to key in dictionary
print(d['jesse'])
```

```python
#loop over the key:value pairs of dictionary
for key, val in d.items():
  print(f'key {key}: val {val}')
```
